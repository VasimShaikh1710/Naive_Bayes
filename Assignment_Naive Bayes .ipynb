```python
Y = df_train['Salary']
Y
```

```python
X = df_train.drop(['Salary'], axis=1)
X
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
list = ['workclass', 'education', 'maritalstatus', 'occupation', 'relationship', 'race', 'sex', 'native']

for i in list:
    ohe = OneHotEncoder(drop = 'first', sparse = False)
    df_train[i] = ohe.fit_transform(df_train[[i]])
```

```python
X = df_train.drop(['Salary'], axis=1)
X
```

```python
# Splitting Train and Test
from sklearn.model_selection._split import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.7)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.naive_bayes import MultinomialNB
MNB = MultinomialNB()
MNB.fit(X_train,Y_train)
```

```python
Y_pred_train = MNB.predict(X_train)
Y_pred_test = MNB.predict(X_test)
Y_pred_test
```

```python
from sklearn.metrics import accuracy_score
print("Training Accuracy Error: ",accuracy_score(Y_train,Y_pred_train).round(2))
print("Test Accuracy Error: ",accuracy_score(Y_test,Y_pred_test).round(2))
```
